```python
import os
from keras.models import load_model
from PIL import Image 
from matplotlib import pyplot as plt
import cv2

import numpy as np

```

```python

model = load_model('/content/drive/My Drive/Image Quality/modelquality.h5')

model.compile(loss='binary_crossentropy',
             optimizer='rmsprop',
              metrics=['accuracy'])

img = cv2.imread('/content/drive/My Drive/Image Quality/test_images/32.png')

img1 = cv2.resize(img,(256,256))

img = np.reshape(img1,[1,256,256,3])

classes = model.predict(img)
```

```python

print ("class is " , classes )
plt.imshow(img1, interpolation='nearest')
plt.show()
```
